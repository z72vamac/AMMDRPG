```python
import pickle as pickle
import data
import json
from itertools import combinations

instances = pickle.load(open("instances.pickle", "rb"))
```

```python
instances[(0, 5, True, 20, 1)].instances[4].A
```

```python
for instance in range(0, 5):
    for size in [5, 10]:
        data = instances[(instance, size, True, 20, 1)]

        dictionary = {}

        with open('instances/AMMDRPG_{a}_{b}.json'.format(a=size, b=instance), 'w') as outfile:
            
            dictionary['orig'] = data.origin
            dictionary['dest'] = data.destination

            G = {}

            G['V'] = {}
            G['E'] = {}
            G['alpha'] = {}

            for g in range(size):
                graph = data.instances[g]
                for v in range(graph.points_number):
                    G['V'][(g+1, v)] = (graph.V[v][0], graph.V[v][1])
                
                for v, w in combinations(range(graph.points_number), 2):
                    if graph.A[v, w] > 0:
                        G['E'][(g+1, v, w)] = graph.A[v, w]
                
                G['alpha'][g+1] = graph.alpha

            dictionary['G'] = G

            dictionary['vM'] = data.truck_speed
            dictionary['vD'] = data.drone_speed

            dictionary['N'] = [20, 30, 40, 50, 60]

            dictionary['fleet_size'] = [1, 2, 3]

            
```
